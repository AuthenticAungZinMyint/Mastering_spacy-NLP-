```python
from collections import Counter
import spacy

nlp = spacy.load("en_core_web_md")
corpus = open("data/atis_utterances.txt", 'r')
corpus = corpus.read().split("\n")
```

```python
all_ent_labels = []
for sentence in corpus:
    doc = nlp(sentence.strip())
    ents = doc.ents
    all_ent_labels += [ent.label_ for ent in ents]
    c = Counter(all_ent_labels)
    print(c)
```

```python
#extract location entities by spacy matcher
import spacy
from spacy.matcher import Matcher

nlp = spacy.load('en_core_web_md')

matcher = Matcher(nlp.vocab)
pattern = [{"POS": "ADP"}, {"ENT_TYPE": "GPE"}]
matcher.add("prepositionLocation", [pattern])

filename = "data/atis_utterance.txt"
file = open(filename, "r", encoding="utf-8")
text = file.read()
doc = nlp(text)
matches = matcher(doc)
for mid, start, end in matches:
    print(doc[start:end])
```

```python
#extract airline org and company name
import spacy
from spacy.matcher import Matcher

nlp = spacy.load('en_core_web_md')

matcher = Matcher(nlp.vocab)
pattern = [{"ENT_TYPE": "ORG", "OP": "+"}]
matcher.add("airlineCompanyName", [pattern])
filename = "data/atis_utterance.txt"
file = open(filename, "r", encoding="utf-8")
text = file.read()
doc = nlp(text)
matches = matcher(doc)
for mid, start, end in matches:
    print(doc[start:end])
```

```python
#extract dates and times 
import spacy
from spacy.matcher import Matcher
from spacy import displacy

nlp = spacy.load("en_core_web_md")

filename = "data/atis_utterance.txt"
file = open(filename, 'r', encoding='UTF-8')
text = file.read()
doc = nlp(text)

matcher = Matcher(nlp.vocab)
pattern = [{"ENT_TYPE": "DATE"}, {"ENT_TYPE": "TIME"}]
matcher.add("timeDate", [pattern])
matches = matcher(doc)
for mid, start, end in matches:
    print(doc[start:end])
```

```python
#display
#colors = {"DATE": "linear-gradient(90deg, #aa9cfc, #fc9ce7)", "TIME": "linear-gradient(90deg, #aa9cfc, #fc9ce7)"}
options = {"ents": ["DATE", "TIME"]}
displacy.serve(doc, style="ent", options=options)
```

```python
#extract abbreviation type entities
import spacy
from spacy.matcher import Matcher
from spacy import displacy

nlp = spacy.load("en_core_web_md")
matcher = Matcher(nlp.vocab)

pattern1 = [{"TEXT": {"REGEX": "\w{1,2}\d{1,2}"}}]
pattern2 = [{"SHAPE": { "IN": ["x", "xx"]}}, 
            {"SHAPE": {"IN": ["d", "dd"]}}]
pattern3 = [{"TEXT": {"IN": ["class", "code",
                            "abbrev", "abbreviation"]}},
           {"SHAPE": {"IN": ["x", "xx"]}}]
pattern4 = [{"POS": "NOUN", "SHAPE": { "IN": ["x", "xx"]}}]

matcher.add("abbrevEntities", [pattern1, pattern2, pattern3, pattern4])

sentences = [
    'what does restriction ap 57 mean',
    'what does the abbreviation co mean',
    'what does fare code qo mean',
    'what is the abbreviation d10',
    'what does code y mean',
    'what does the fare code f and fn mean',
    'what is booking class c'
]
for sent in sentences:
    doc = nlp(sent)
    matches = matcher(doc)
    for mid, start, end in matches:
        print(doc[start:end])
```
